```python
#importing numpy pandas mathplotand other methods
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#first read the data file
df= pd.read_csv('heart_failure_clinical_records.csv')
```

```python
# quick look to dataset
df.head(33)
```

```python
#Print summary statistics
df.describe()
```

```python
#the complete information about the dataset
df.info()
```

```python
# Inspect missing values in the dataset
print(df.isnull().values.sum())

# Replace the ' 's with NaN
df = df.replace(" ",np.NaN)


# Impute the missing values with mean imputation
df = df.fillna(df.mean())

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

```python
# seaborn histogram 
sns.distplot(df['age'], hist=True, kde=False, 
             bins=12, color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Age count')
plt.xlabel('Age')

plt.ylabel('Count')
```

In The Above seaborn distplot, it is seen that
- The minimum age in the dataset is from 40 yrs to almost 100 yrs old .
- People from 50 years to 70 years have suffered from heart failure the most.Especially, people of age 50 and 60 have suffered the most.
- After the 70s, there is huge decrease in the disease count.

```python
# seaborn Scatterplot
sns.scatterplot(x=df['age'], y=df['anaemia'])
```

In this scatterplot, the interaction of age with anaemia is observed.Anaemia count raises from 0 to 1 for the age of people from 40 to 100 years old.
This scatterplot simply show that people of age 50-60 years have the highest anaemia count and as people get older the disease ratio decreases.

```python
#creatinine_phosphokinase
# seaborn Scatterplot
sns.scatterplot(x=df['creatinine_phosphokinase'], y=df['age'])
```

Similarly, In this scatterplot, We can see that the highest creatinine phosphokinase count is 8000 of a person of age around 55 years. And, almost all the age of people have their creatinine phosphokinase count around 0 to 1000s in average.

```python
# seaborn Scatterplot
sns.scatterplot(x=df['platelets'], y=df['age'])
```

Here, Relation of platelets and age is observed. People from their 40s to late 70s have the most platelets and the average no of platelets ranges from 2000000 to 400000.

```python
#this is heat map pearson correlation matrix
corrmat = df.corr()
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size = 15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df.corr(),
            cmap = colormap,
            square = True,
            annot = True,
            linewidths=0.1,vmax=1.0, linecolor='white',
            annot_kws={'fontsize':12 })
plt.show()
```

In The Heat-Map OR Correlation map, in general,

Each square shows the correlation between the variables on each axis. Correlation ranges from -1 to +1. Values closer to zero means there is no linear trend between the two variables. The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increases. The diagonals are all 1/dark green because those squares are correlating each variable to itself (so it's a perfect correlation). For the rest the larger the number and darker the color the higher the correlation between the two variables.
Above,
- For example, Time and Death_Event have perfect correlation.
- As the color darkens , the correlation tends to negative and their relation is inverse. One's increase results in decrement of another.
- As the color lightens, the correlation is towards positive and they have mutual increase or decrease .
